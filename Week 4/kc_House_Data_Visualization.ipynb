```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, IntSlider
```

Berikut adalah ringkasan umum untuk masing-masing kolom dalam DataFrame `df`:

1. **id**: Ini mungkin adalah ID unik yang diberikan untuk setiap properti. Ini adalah cara yang baik untuk mengidentifikasi properti secara unik.

2. **date**: Kolom ini mungkin berisi tanggal penjualan atau transaksi properti. Ini dapat digunakan untuk analisis tren seiring waktu.

3. **price**: Ini adalah harga penjualan properti. Ini adalah kolom yang sangat penting untuk analisis properti.

4. **bedrooms**: Menunjukkan jumlah kamar tidur dalam properti.

5. **bathrooms**: Menunjukkan jumlah kamar mandi dalam properti.

6. **sqft_living**: Ini mungkin adalah luas ruang hidup dalam kaki persegi. Ini adalah ukuran penting dalam mengevaluasi ukuran properti.

7. **sqft_lot**: Ini mungkin adalah ukuran total tanah dalam kaki persegi yang dimiliki oleh properti.

8. **floors**: Menunjukkan jumlah lantai dalam bangunan properti.

9. **waterfront**: Ini mungkin adalah indikator apakah properti tersebut berada di tepi air (1 untuk ya, 0 untuk tidak).

10. **view**: Menunjukkan sejauh mana properti memiliki pemandangan (mungkin dalam skala 0-4).

11. **condition**: Menunjukkan kondisi umum properti (mungkin dalam skala 1-5).

12. **grade**: Ini mungkin adalah peringkat kualitas bangunan dan desain properti (mungkin dalam skala 1-13).

13. **sqft_above**: Ini mungkin adalah luas lantai atas dalam faki persegi.

14. **sqft_basement**: Ini mungkin adalah luas ruang bawah tanah dalam faki persegi.

15. **yr_built**: Tahun di mana bangunan properti dibangun.

16. **yr_renovated**: Tahun renovasi terakhir properti, jika ada.

17. **zipcode**: Kode pos tempat properti berada.

18. **lat**: Koordinat lintang lokasi properti.

19. **long**: Koordinat bujur lokasi properti.

20. **sqft_living15**: Luas ruang hidup rata-rata dari 15 properti terdekat dalam faki persegi.

21. **sqft_lot15**: Ukuran total tanah rata-rata dari 15 properti terdekat dalam faki persegi.

prompt: visualisasikan distribusi masing-masing fitur

```python
# Visualisasi distribusi untuk setiap fitur
for column in df.columns:
    if df[column].dtype == 'float64':
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], kde=True, color='cyan')
        plt.title(f'Distribusi {column}', size=15)
        plt.xlabel(column, size=12)
        plt.ylabel('Frequency', size=12)
        plt.show()
```

```python
# Visualisasi distribusi untuk setiap fitur
for column in df.columns:
    if df[column].dtype == 'float64':
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], kde=True, color='orange')
        plt.title(f'Distribusi {column}', size=15)
        plt.xlabel(column, size=12)
        plt.ylabel('Frequency', size=12)
        plt.show()
```

ubah visualisasi untuk 3 baris 3 kolom

```python
# Membuat subplot 3x3
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.subplots_adjust(wspace=0.5, hspace=0.5)  # Menambahkan jarak antar subplot

# Visualisasi distribusi untuk setiap fitur
numeric_columns = df.select_dtypes(include=['float64']).columns
for i, column in enumerate(numeric_columns):
    row, col = divmod(i, 3)  # Menghitung posisi subplot
    ax = axes[row, col]

    sns.histplot(df[column], kde=True, color='orange', ax=ax)
    ax.set_title(f'Distribusi {column}', size=12)
    ax.set_xlabel(column, size=10)
    ax.set_ylabel('Frequency', size=10)

# Menghilangkan subplot yang tidak digunakan
for i in range(len(numeric_columns), 9):
    row, col = divmod(i, 3)
    fig.delaxes(axes[row, col])

plt.show()
```

```python
# Fungsi untuk memplot distribusi dengan bin width interaktif
def plot_distribution(feature, bin_width):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], bins=bin_width, kde=True, color='cyan')
    plt.title(f'Distribusi {feature} (Bin Width: {bin_width})', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Frequency', size=12)
    plt.show()

# Interaksi untuk memilih fitur dan bin width
interact(plot_distribution, feature=df.columns, bin_width=IntSlider(min=1, max=50, step=1, value=10))
```

```python
# Visualisasi boxplot untuk setiap fitur
for column in df.columns:
    if df[column].dtype == 'float64':
        plt.figure(figsize=(8, 6))
        sns.boxplot(y=df[column], color='khaki')
        plt.title(f'Boxplot {column}', size=15)
        plt.ylabel(column, size=12)
        plt.show()
```

prompt: ubah 3 baris 3 kolom

```python
# Mengatur ukuran dan tata letak subplot
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Visualisasi boxplot untuk setiap fitur dalam tata letak 3x3
numeric_columns = df.select_dtypes(include=['float64']).columns
for i, column in enumerate(numeric_columns[:9]):  # Hanya mengambil 9 kolom pertama
    sns.boxplot(x=df[column], color='green', orient='h', ax=axes[i//3, i%3])
    axes[i//3, i%3].set_title(f'Boxplot {column}', size=12)
    axes[i//3, i%3].set_xlabel(column, size=10)
    axes[i//3, i%3].set_ylabel('Value', size=10)

# Menghilangkan subplot yang tidak digunakan
for i in range(len(numeric_columns), 9):
    row, col = divmod(i, 3)
    fig.delaxes(axes[row, col])

# Mengatur tata letak secara rapi
plt.tight_layout()
plt.show()
```

```python
# Membuat heatmap untuk setiap fitur
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Fitur', size=15)
plt.show()
```

```python
# Mengatur tata letak secara rapi
plt.figure(figsize=(12, 8))

# Membuat violin chart untuk setiap fitur
numeric_columns = df.select_dtypes(include=['float64']).columns
for i, column in enumerate(numeric_columns[:9]):  # Hanya mengambil 9 kolom pertama
    plt.subplot(3, 3, i+1)
    sns.violinplot(y=df[column], color='skyblue')
    plt.title(f'Violin Chart {column}', size=12)
    plt.ylabel(column, size=10)

# Mengatur tata letak secara rapi
plt.tight_layout()
plt.show()
```
